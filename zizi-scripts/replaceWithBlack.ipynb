# Fades

```python
import os
import glob
import numpy as np
import cv2
from PIL import Image, ImageEnhance


# cv2.imwrite(out+"/"+os.path.basename(directory)+"/"+f+".png", emptyFig)

fc = "full"
    

emptyFig = np.zeros((832,832,3), np.uint8)

outs=[]

foundFirst = False
foundLast = False
Last = True
first=1000

for out in sorted(glob.glob("../zizi-data/results/*-"+fc+"2")):
    outs.append(out)
    
print(outs)

def makeFade(number, direction, directory):
    print("fade "+str(number)+direction)
    if direction == "forward":
        print("fade from "+str(int(number)) + " to " +str(int(number)+25))
        
        fadenum=0.0

        for f in range(0,25):
            im = Image.open(out+'/'+os.path.basename(directory)+'/'+str(int(number)+(f)).zfill(6)+'.png')
            enhancer = ImageEnhance.Brightness(im)
            im = enhancer.enhance(fadenum)
        #     save(im)
#             display(im)
            im.save(out+'/'+os.path.basename(directory)+'/'+str(int(number)+(f)).zfill(6)+'.png')
            fadenum = fadenum + 1.0/25
            print(str(fadenum) + " " +out+'/'+os.path.basename(directory)+'/'+str(int(number)+(f)).zfill(6)+'.png')
        
    if direction == "backward":
        print("fade from "+str(int(number)) + " to " +str(int(number)-25))
        
        fadenum=0.0

        for f in range(0,25):
            im = Image.open(out+'/'+os.path.basename(directory)+'/'+str(int(number)-int(f)).zfill(6)+'.png')
            enhancer = ImageEnhance.Brightness(im)
            im = enhancer.enhance(fadenum)
        #     save(im)
#             display(im)
            im.save(out+'/'+os.path.basename(directory)+'/'+str(int(number)-int(f)).zfill(6)+'.png')
            fadenum = fadenum + 1.0/25
            print(str(fadenum) + " " +out+'/'+os.path.basename(directory)+'/'+str(int(number)+int(f)).zfill(6)+'.png')
        



for directory in sorted(glob.glob("../zizi-data/datasets/acts/"+fc+"/test_densepose/*")):
# directory = "../zizi-data/datasets/acts/full/test_densepose/fiveyears-ruby-full/"
    print(directory)
    for file in sorted(glob.glob(directory+"/*")):
        f = os.path.basename(os.path.splitext(file)[0])[:6]
        b = os.path.getsize(file)
        c = os.path.getsize("../zizi-data/datasets/acts/"+fc+"/test_openpose/"+os.path.basename(directory)+"/"+f+"_keypoints.json")

        if os.path.basename(directory) == "iam-me-"+fc:
            denseLim = 61396
        else:
            denseLim = 16887 

        if b != denseLim or c != 27:
            if foundFirst==False:
                first = int(f)
                print(f)
                foundFirst = True
                makeFade(f, "forward", directory)

        if b == denseLim and c == 27 and int(f) > 2000:
            if foundLast==False:
                last = int(f)
                print(f)
                foundLast = True
                makeFade(f, "backward", directory)
                print("_____VID DONE     "+os.path.basename(directory))

    foundFirst = False
    foundLast = False
    
    
    
```
