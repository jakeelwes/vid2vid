```python
import os
import glob
import numpy as np
import cv2
from PIL import Image, ImageEnhance
```

```python
fc = "close"
name="*"
# name="*"
# Not Amalgam, Lilly or Pussi
 
emptyFig = np.zeros((832,832,3), np.uint8)

outs=[]

for out in sorted(glob.glob("../zizi-data/results/"+name+"-"+fc)):
    outs.append(out)

print(len(outs))
print(outs)
```

```python
def makeFade(number, direction, directory, outdir):
    print("fade "+str(number)+direction)
    if direction == "forward":
        print("fade from "+str(int(number)) + " to " +str(int(number)+12))

        fadenum=0.0

        for f in range(0,12):
#             try:
            im = Image.open(outdir+'/'+os.path.basename(directory)+'/'+str(int(number)+(f)).zfill(6)+'.jpg')
            enhancer = ImageEnhance.Brightness(im)
            im = enhancer.enhance(fadenum)
        #     save(im)
#             display(im)
            im.save(outdir+'/'+os.path.basename(directory)+'/'+str(int(number)+(f)).zfill(6)+'.jpg')
            fadenum = fadenum + 1.0/12
            print(str(fadenum) + " " +outdir+'/'+os.path.basename(directory)+'/'+str(int(number)+(f)).zfill(6)+'.jpg')
#             except OSError as e:
#                 pass
            
    if direction == "backward":
        print("fade from "+str(int(number)) + " to " +str(int(number)-12))
        
        fadenum=0.0

        for f in range(0,12):
            im = Image.open(outdir+'/'+os.path.basename(directory)+'/'+str(int(number)-int(f)).zfill(6)+'.jpg')
            enhancer = ImageEnhance.Brightness(im)
            im = enhancer.enhance(fadenum)
        #     save(im)
#             display(im)
            im.save(outdir+'/'+os.path.basename(directory)+'/'+str(int(number)-int(f)).zfill(6)+'.jpg')
            fadenum = fadenum + 1.0/12
            print(str(fadenum) + " " +outdir+'/'+os.path.basename(directory)+'/'+str(int(number)+int(f)).zfill(6)+'.jpg')
        
```

```python
for directory in sorted(glob.glob("../zizi-data/datasets/acts/"+fc+"/test_densepose/*")):
    if os.path.basename(directory) != "aintro-"+fc:
        print(directory)
        
        files=[]

        for file in sorted(glob.glob(directory+"/*")):
            files.append(file)

        lastFrame=len(files)
        first=0
        last=0

        if os.path.basename(directory) == "iam-me-"+fc:
            denseLim = 61396
        else:
            denseLim = 16887
            
        for out in outs:
            for n in range(0,500):
                f = os.path.basename(os.path.splitext(files[n])[0])[:6]
                b = os.path.getsize(files[n])
                c = os.path.getsize("../zizi-data/datasets/acts/"+fc+"/test_openpose/"+os.path.basename(directory)+"/"+f+"_keypoints.json")
                if b == denseLim and c == 27:
                    print(out+"/"+os.path.basename(directory)+"/"+f+".jpg" + " - writing black")
                    cv2.imwrite(out+"/"+os.path.basename(directory)+"/"+f+".jpg", emptyFig)
                    first=n

            for n in range(lastFrame-1,lastFrame-500,-1):
                f = os.path.basename(os.path.splitext(files[n])[0])[:6]
                b = os.path.getsize(files[n])
                c = os.path.getsize("../zizi-data/datasets/acts/"+fc+"/test_openpose/"+os.path.basename(directory)+"/"+f+"_keypoints.json")
                if b == denseLim and c == 27:
                    print(out+"/"+os.path.basename(directory)+"/"+f+".jpg" + " - writing black")
                    cv2.imwrite(out+"/"+os.path.basename(directory)+"/"+f+".jpg", emptyFig)
                    last=n

            print(str(first) + ' ' + str(last) + ' ' + str(lastFrame))
            print(out+"/"+os.path.basename(directory)+"/"+str(last)+".jpg" + " - writing black")
            makeFade(first, "forward", directory, out)
            makeFade(last, "backward", directory, out)
    
        print(str(first) + ' ' + str(last) + ' ' + str(lastFrame))
        print("________ DONE SONG ________")


```
